# `Train Test split`

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = df.drop(['y'], axis=1)
y = df['y']
X.shape, y.shape
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
x_train, x_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.15, random_state=42)
x_train.shape, y_train.shape, x_validation.shape, y_validation.shape, x_test.shape, y_test.shape
```

# `Normalization`

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_validation = scaler.transform(x_validation)
x_test = scaler.transform(x_test)
```

# `Model`

```python
from sklearn.ensemble import *
from sklearn.tree import *
```

### `Bagging Classifier`

```python
model_bagging = BaggingClassifier(estimator= DecisionTreeClassifier(max_depth=2, random_state=42, ccp_alpha=0.01),
                                n_estimators=10, max_samples=0.7, oob_score=True, random_state=42)
model_bagging.fit(x_train, y_train)
```

```python
print(model_bagging.oob_score_)
print(model_bagging.score(x_validation, y_validation))
```

```python
print(model_bagging.score(x_test, y_test))
```

### `Random Forest`

```python
model_randomforest = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=7, random_state=42, oob_score=True)
model_randomforest.fit(x_train, y_train)
```

```python
print(model_randomforest.oob_score_)
print(model_randomforest.score(x_validation, y_validation))
```

```python
print(model_randomforest.score(x_test, y_test))
```

### `Ada Boost`

```python
model_ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1,max_leaf_nodes=2, random_state=42, ccp_alpha=0.01),
                                n_estimators=200,learning_rate=0.1,algorithm='SAMME' ,random_state=42)

model_ada.fit(x_train, y_train)
```

```python
print(model_ada.score(x_train, y_train))
print(model_ada.score(x_validation, y_validation))
```

```python
print(model_ada.score(x_test, y_test))
```

### `Logistic Regression`

```python
from sklearn.linear_model import SGDClassifier
```

```python
model_SGDclassifier = SGDClassifier(penalty=None, random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

# `Stacking`

```python
model = StackingClassifier(estimators=[
                                        ('Bagging', model_bagging),
                                        ('Random_Forest', model_randomforest),
                                        ('Ada_Boost', model_ada)
                                        ],
                                         final_estimator = model_SGDclassifier 
                            )
                            
model.fit(x_train, y_train)
```

```python
from IPython.display import Image
Image(filename=r'Stacking Classifier.PNG')
```

```python
print(model.score(x_train, y_train))
print(model.score(x_validation, y_validation))
```

```python
print(model.score(x_test, y_test))
```

```python
print('Bagging model Test Score:       ', model_bagging.score(x_test, y_test))
print('Random model forest Test Score: ', model_randomforest.score(x_test, y_test))
print('Ada boost model Test Score:     ', model_ada.score(x_test, y_test))
print('Stacking model Test Score:      ', model.score(x_test, y_test))
```
